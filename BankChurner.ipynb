Tỷ lệ oulier thấp trong mỗi biến. Hơn nữa chúng không biểu thị thông tin không nhất quán như lỗi đo lường, nên sẽ không xử lý oulier.

Xử lý outlier ở 2 biến liên tục có rất nhiều oulier là Total_Amt_Chng_Q4_Q1, Total_Ct_Chng_Q4_Q1

```python
p25 = df['Total_Amt_Chng_Q4_Q1'].quantile(0.25)
p75 = df['Total_Amt_Chng_Q4_Q1'].quantile(0.75)
iqr = p75 - p25

lower_bound = p25 - 1.5 * iqr
upper_bound = p75 + 1.5 * iqr

df = df[(df['Total_Amt_Chng_Q4_Q1'] >= lower_bound) & (df['Total_Amt_Chng_Q4_Q1'] <= upper_bound)]
df = df[(df['Total_Ct_Chng_Q4_Q1'] >= lower_bound) & (df['Total_Ct_Chng_Q4_Q1'] <= upper_bound)]
```

```python
# xóa unknown khỏi các cọt education, marital, income
df = df[df['Education_Level'] != 'Unknown']
df = df[df['Marital_Status'] != 'Unknown']
df = df[df['Income_Category'] != 'Unknown']
```

**ex1 smote + minmax + log**

```python
x = df_dummy.drop('Attrition_Flag', axis=1)
y = df_dummy['Attrition_Flag']

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

sm = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = sm.fit_resample(X_train, y_train)

scale_columns = ['Customer_Age', 'Dependent_count', 'Months_on_book', 'Total_Relationship_Count',
                 'Months_Inactive_12_mon', 'Contacts_Count_12_mon', 'Credit_Limit', 'Total_Revolving_Bal',
                 'Avg_Open_To_Buy', 'Total_Amt_Chng_Q4_Q1', 'Total_Trans_Amt', 'Total_Trans_Ct', 'Total_Ct_Chng_Q4_Q1',
                 'Avg_Utilization_Ratio']

scaler = MinMaxScaler()
X_train_resampled[scale_columns] = scaler.fit_transform(X_train_resampled[scale_columns])
X_test[scale_columns] = scaler.transform(X_test[scale_columns])
```

```python
model_log = LogisticRegression()
model_log.fit(X_train_resampled, y_train_resampled)
y_pred = model_log.predict(X_test)

print(classification_report(y_test, y_pred))

ConfusionMatrixDisplay.from_estimator(model_log, X_test, y_test)
```

```python
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

x = df_dummy.drop('Attrition_Flag', axis=1)
y = df_dummy['Attrition_Flag']

X_train_val, X_test, y_train_val, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
X_train, X_val, y_train, y_val = train_test_split(X_train_val, y_train_val, test_size=0.25, random_state=42)

xgb = XGBClassifier(learning_rate=0.1, n_estimators=300, max_depth=5)
xgb.fit(X_train, y_train)

y_pred_train = xgb.predict(X_train)
y_pred_val = xgb.predict(X_val)

print("Kết quả trên tập train:")
print(classification_report(y_train, y_pred_train))

print("Kết quả trên tập validation:")
print(classification_report(y_val, y_pred_val))

y_pred_test = xgb.predict(X_test)
print("Kết quả trên tập test:")
print(classification_report(y_test, y_pred_test))
```
